# Python LOOPS

### Q1)

```python
num = 1

while num < 6:
    print(num)
    num += 1
```

### Q2)

```python
sum = 0
num = 1
while num < 11:
    sum += num
    num += 1

print(sum)
```

### Q3) 

```python
num = 5
fac = 1

for i in range(1,num+1):
    fac *= i

print(f"Factorial of {num} is {fac}")
```

### Q4)

```python
string = "hello this is a sample string needed for calculation of vowels"
num_of_vowels = 0

for ch in string:
    if ch in "aeiou":
        num_of_vowels += 1

print("Number Of Vowels:",num_of_vowels)
```

### Q5)

```python
for i in range(1,6):
    curr_string = ""
    for j in range(i):
        curr_string += "*"
    
    print(curr_string)
```

### Q6)

```python
num = 2

for i in range(1,num+1):
    print(f"\nTable of {i}")
    for j in range(1,11):
        print(f"{i}X{j} = {i*j}")
```
