### Q7)

```python
#Palindromic Number from 1 to n using generator function.
def generate_palindromes(n):
    for num in range(1, n + 1):
        if str(num) == str(num)[::-1]:
            yield num

n = 200
palindromes = generate_palindromes(n)

for palindrome in palindromes:
    print(palindrome)
```

### Q10)

```python
#Generator to generate prime numbers till n:

def prime_generator(n):
    for num in range(2,n+1):
        prime = True
        for i in range(2,num):
            if num % i == 0:
                prime = False
        
        if prime:
            yield num


primes = prime_generator(1000)

for num in primes:
    print(num)
```

### Q39)

```python
# Abstract VS Normal Methods

answer = """
Abstract methods and normal methods in programming, especially within the context of object-oriented 
programming, serve different purposes and have distinct characteristics:

1. **Definition and Purpose**:
   - **Abstract Methods**: An abstract method is a method that is declared in an abstract class (or 
   interface) but does not contain any implementation. It serves as a blueprint for derived classes, 
   enforcing that they must provide their own implementation of the method.
   - **Normal Methods**: A normal method is a method that has a complete implementation within a class. 
   It defines specific behavior that objects of the class can perform.

2. **Usage**:
   - **Abstract Methods**: These methods are used when you want to ensure that certain methods are 
   implemented in all subclasses of an abstract class. They are essential for defining a common interface 
   for a group of related classes.
   - **Normal Methods**: These methods are used to define the actual behavior and operations that an 
   object can perform. They encapsulate functionality that can be directly used by creating an instance 
   of the class.

3. **Implementation**:
   - **Abstract Methods**: Cannot have any implementation in the abstract class. They are declared 
   with the `abstract` keyword in many programming languages (e.g., `@abstractmethod` decorator in Python).
   - **Normal Methods**: Have a complete implementation within the class where they are defined, including 
   the method body with specific logic and behavior.

4. **Instantiation**:
   - **Abstract Methods**: Classes containing abstract methods (abstract classes) cannot be instantiated 
   directly. They must be subclassed, and the abstract methods must be implemented in the derived classes 
   before objects can be created.
   - **Normal Methods**: Classes with only normal methods can be instantiated directly, and the methods 
   can be called on the instances of the class.

5. **Enforcement**:
   - **Abstract Methods**: Serve as a contractual obligation for subclasses. Any subclass of an abstract 
   class must provide concrete implementations for all abstract methods.
   - **Normal Methods**: Do not impose any requirements on subclasses. They provide specific functionality 
   that can be inherited or overridden by subclasses as needed.

In summary, abstract methods are used to define a required interface without implementation, ensuring that 
subclasses adhere to a certain contract, while normal methods provide concrete behavior that can be 
directly executed by instances of a class.
"""
```

### Q41)

```python
# Example of abstract

from abc import ABC, abstractmethod

# Define an abstract base class (interface)
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

# Implement the interface in a subclass
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

# Another subclass implementing the interface
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

    def perimeter(self):
        return 2 * 3.14 * self.radius

# Trying to instantiate Shape will raise an error
# shape = Shape()  # This will raise a TypeError

# Correctly instantiating the concrete subclasses
rectangle = Rectangle(4, 5)
circle = Circle(3)

print(f"Rectangle area: {rectangle.area()}, perimeter: {rectangle.perimeter()}")
print(f"Circle area: {circle.area()}, perimeter: {circle.perimeter()}")
```

### Q47)

```python
# Achieving Encapsulation

answer = """
To achieve encapsulation in Python:

1. **Class Definition**: Define a class with attributes and methods that manipulate those attributes.
   
2. **Access Control**: Use naming conventions to indicate visibility:
   - Public attributes/methods have no leading underscores (`self.attribute` or `def method(self)`).
   - Protected attributes/methods have a single leading underscore (`self._attribute` or 
   `def _method(self)`).
   - Private attributes/methods have a double leading underscore (`self.__attribute` or 
   `def __method(self)`), name-mangled to prevent direct access.

3. **Properties and Methods**: Use getter and setter methods (`@property` decorator) to control access to 
private attributes, ensuring data integrity and encapsulation. This approach promotes modular and 
maintainable code by hiding implementation details and exposing only necessary functionality.
"""
```

### Q48)

```python
# Encapsulation By Passing

answer = """
In Python, encapsulation can be bypassed in several ways due to its dynamic and flexible nature:

1. **Name Mangling**: While attributes with double underscores (`__attribute`) are intended to be private 
and are name-mangled, Python does not enforce strict access restrictions. They can still be accessed using 
mangled names (`_ClassName__attribute`), although this is discouraged.

2. **Direct Access**: All attributes and methods can technically be accessed directly from outside the 
class instance in Python, regardless of their intended visibility (public, protected, or private). This 
is because Python does not have strict access modifiers like some other languages.

3. **Monkey Patching**: Python allows classes and objects to be modified dynamically at runtime. This 
means that attributes and methods can be added, modified, or deleted from outside the class definition, 
potentially altering the behavior intended by encapsulation.

4. **Using `__dict__`**: Every Python object has a `__dict__` attribute that stores its attributes. This 
allows direct access to and modification of attributes, including private ones, bypassing encapsulation 
mechanisms.

5. **Reflection and Introspection**: Python's introspection capabilities (`getattr()`, `setattr()`, 
`hasattr()`, etc.) allow inspection and modification of attributes and methods at runtime, which can 
be used to bypass encapsulation.

Despite these possibilities, it is generally recommended to follow encapsulation principles and conventions 
in Python to ensure code clarity, maintainability, and to minimize unintended side effects from direct 
manipulation of class internals.
"""
```

### Q49)

```python
# Bank Account class

class BankAccount:
    def __init__(self,balance):
        self.__balance = balance

    def withdraw(self,amt):
        if self.__balance > amt:
            self.__balance -= amt
            return True
        else:
            return False


    def deposit(self,amt):
        if amt > 0:
            self.__balance += amt
            return True
        else:
            return False

    def check_balance(self):
        print(f"Your Balance: {self.__balance}")        


my_acc = BankAccount(1000)

my_acc.check_balance()
my_acc.deposit(100000)

my_acc.check_balance()
my_acc.withdraw(5000)

my_acc.check_balance()
```

### Q53)

```python
#Decorator with arguments and printing function name

def decorator(func):
    def wrapper(*args, **kwargs):
        print(f"Statement Before Function '{func.__name__}'")
        result = func(*args, **kwargs)
        print(f"Statement After Function '{func.__name__}'")
        return result

    return wrapper

@decorator
def func():
    print("Hello World!!!")

@decorator
def another_func(name):
    print(f"Hello, {name}!")

func()
another_func("Alice")
```

### Q55)

```python
# Decorator Passing Arguments

def decorator_with_args(pre_message, post_message):
    def decorator(func):
        def wrapper(*args, **kwargs):
            print(f"{pre_message}")
            result = func(*args, **kwargs)
            print(f"{post_message}")
            return result
        return wrapper
    return decorator

@decorator_with_args("Statement Before Function", "Statement After Function")
def func():
    print("Hello World!!!")

@decorator_with_args("Start Function", "End Function")
def another_func(name):
    print(f"Hello, {name}!")

func()
another_func("Alice")
```

### Q64)

```python
# FOR __add__

answer = """
The __add__ method in Python is a special method used to define the behavior of the + operator for instances 
of a class. By overriding __add__, you can specify how two objects of your custom class should be added 
together. This is particularly useful for classes that represent mathematical constructs or collections, 
where addition operations make sense.
"""

#Example
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        return NotImplemented

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

vector1 = Vector(2, 3)
vector2 = Vector(4, 5)

result = vector1 + vector2

print(result)
```

### Q66)

```python
# For iter next

answer = """
The __iter__ and __next__ methods in Python are used together to define an iterator for instances of a class.
Iterators are objects that implement the iterator protocol, which consists of these two methods:

__iter__: This method returns the iterator object itself. It is called when the iterator is initialized. 
Typically, it just returns self.

__next__: This method returns the next item in the sequence. It is called on each iteration to fetch the 
next item. When there are no more items to return, it should raise a StopIteration exception.
"""

#Example
class Counter:
    def __init__(self, low, high):
        self.current = low
        self.high = high

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.high:
            raise StopIteration
        else:
            self.current += 1
            return self.current - 1

counter = Counter(1, 5)

for num in counter:
    print(num)
```

### Q68)

```python
#Setter
answer = """
In Python, a setter method is used to modify the value of a private attribute of a class in a controlled 
manner. It allows you to encapsulate the logic for setting the value of an attribute, enabling validation, 
computation, or side effects when assigning a new value.
"""

#Example
class Circle:
    def __init__(self, radius):
        self._radius = radius 

    def get_radius(self):
        return self._radius

    def set_radius(self, new_radius):
        if new_radius > 0:
            self._radius = new_radius
        else:
            raise ValueError("Radius must be a positive number.")

circle = Circle(5)

print(circle.get_radius())
circle.set_radius(10)
print(circle.get_radius()) 

try:
    circle.set_radius(-1)
except ValueError as e:
    print(e)
```

### Q69)

```python
# Property

answer = """
The @property decorator in Python facilitates the implementation of managed attributes within classes by 
allowing methods to be accessed like attributes. This decorator is essential for encapsulating attribute 
access with custom behavior such as validation, computation, or dynamic retrieval of values. By defining 
getter, setter, and deleter methods decorated with @property, classes can control how their attributes are 
accessed and modified, promoting cleaner and more readable code while maintaining compatibility with 
existing usage patterns that interact with attributes directly. This approach enhances encapsulation, 
simplifies maintenance, and improves the overall design and usability of Python classes.
"""

#Example
class Circle:
    def __init__(self, radius):
        self._radius = radius

    @property
    def radius(self):
        return self._radius

    @radius.setter
    def radius(self, value):
        if value > 0:
            self._radius = value
        else:
            raise ValueError("Radius must be a positive number.")

circle = Circle(5)

print(circle.radius) 

circle.radius = 10
print(circle.radius) 
try:
    circle.radius = -1
except ValueError as e:
    print(e)
```

```python
#Deleter
answer = """
Deleter function is used to delete the function attributes 
"""

#Example
class MyClass:
    def __init__(self):
        self._value = None
    
    @property
    def value(self):
        return self._value
    
    @value.setter
    def value(self, new_value):
        self._value = new_value
    
    @value.deleter
    def value(self):
        del self._value
        print("Value has been deleted.")

obj = MyClass()

obj.value = 42

print(obj.value)

del obj.value  
```
