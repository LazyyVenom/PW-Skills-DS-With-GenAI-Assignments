# Python Break And Continue

### Q1)

```python
for num in range(1,11):
    if num == 5:
        break
    print(num)
```

### Q2)

```python
specific_element = "banana"

fruits = ["apple","orange","lemon","banana","lichi"]

for fruit in fruits:
    if fruit == specific_element:
        break
    print(fruit)
```

### Q3) 

```python
for num in range(1,11):
    if num % 2 == 0:
        continue
    print(num)
```

### Q4)

```python
for i in range(0,10):
    print(i)
```

### Q5)

```python
for i in range(1,6):
    for j in range(1,11):
        if j == 2:
            break
        print(f"{i}X{j}={i*j}")
```

### Q6)

```python
upper_limit = 20
count = 0
while count <= upper_limit:
    if count % 2 == 0:
        count += 1
        continue
    print(count)
    count += 1
```
