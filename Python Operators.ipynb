### Q2)

```python
# Initialize the variable
variable = 50

# Perform addition assignment
variable += 20  # equivalent to variable = variable + 20
print("After addition assignment (+=):", variable)  

# Perform subtraction assignment
variable -= 10  # equivalent to variable = variable - 10
print("After subtraction assignment (-=):", variable)  

# Perform multiplication assignment
variable *= 2  # equivalent to variable = variable * 2
print("After multiplication assignment (*=):", variable)

# Perform division assignment
variable /= 4  # equivalent to variable = variable / 4
print("After division assignment (/=):", variable)

# Perform integer division assignment
variable //= 3  # equivalent to variable = variable // 3
print("After integer division assignment (//=):", variable) 

# Perform modulus assignment
variable %= 3  # equivalent to variable = variable % 3
print("After modulus assignment (%=):", variable) 

# Perform exponentiation assignment
variable **= 4  # equivalent to variable = variable ** 4
print("After exponentiation assignment (**=):", variable)

# Perform bitwise AND assignment
variable = 5  # reset variable to 5 for demonstration
variable &= 3  # equivalent to variable = variable & 3
print("After bitwise AND assignment (&=):", variable) 

# Perform bitwise OR assignment
variable |= 2  # equivalent to variable = variable | 2
print("After bitwise OR assignment (|=):", variable) 

# Perform bitwise XOR assignment
variable ^= 1  # equivalent to variable = variable ^ 1
print("After bitwise XOR assignment (^=):", variable) 

# Perform bitwise left shift assignment
variable <<= 2  # equivalent to variable = variable << 2
print("After bitwise left shift assignment (<<=):", variable)

# Perform bitwise right shift assignment
variable >>= 1  # equivalent to variable = variable >> 1
print("After bitwise right shift assignment (>>=):", variable) 
```

```python
# Define two numbers
num1 = 15
num2 = 20

# Compare the numbers and print the results
print("num1 == num2:", num1 == num2)  # Expected output: False
print("num1 != num2:", num1 != num2)  # Expected output: True
print("num1 < num2:", num1 < num2)    # Expected output: True
print("num1 > num2:", num1 > num2)    # Expected output: False
print("num1 <= num2:", num1 <= num2)  # Expected output: True
print("num1 >= num2:", num1 >= num2)  # Expected output: False
```

### Q6)

```python
num1 = 12  # In binary: 1100
num2 = 5   # In binary: 0101

# Perform bitwise AND
bitwise_and = num1 & num2
print(f"{num1} & {num2} = {bitwise_and}")  # Expected output: 4 (binary: 0100)

# Perform bitwise OR
bitwise_or = num1 | num2
print(f"{num1} | {num2} = {bitwise_or}")  # Expected output: 13 (binary: 1101)

# Perform bitwise XOR
bitwise_xor = num1 ^ num2
print(f"{num1} ^ {num2} = {bitwise_xor}")  # Expected output: 9 (binary: 1001)

# Perform bitwise NOT on num1
bitwise_not_num1 = ~num1
print(f"~{num1} = {bitwise_not_num1}")  # Expected output: -13 (binary: -1101, two's complement form)

# Perform bitwise NOT on num2
bitwise_not_num2 = ~num2
print(f"~{num2} = {bitwise_not_num2}")  # Expected output: -6 (binary: -0110, two's complement form)

# Perform left shift on num1
left_shift_num1 = num1 << 2
print(f"{num1} << 2 = {left_shift_num1}")  # Expected output: 48 (binary: 110000)

# Perform right shift on num1
right_shift_num1 = num1 >> 2
print(f"{num1} >> 2 = {right_shift_num1}")  # Expected output: 3 (binary: 0011)

# Perform left shift on num2
left_shift_num2 = num2 << 2
print(f"{num2} << 2 = {left_shift_num2}")  # Expected output: 20 (binary: 10100)

# Perform right shift on num2
right_shift_num2 = num2 >> 2
print(f"{num2} >> 2 = {right_shift_num2}")  # Expected output: 1 (binary: 0001)
```

### Q8)

```python
num1 = 10
num2 = 20

# Use the ternary operator to determine the larger number
larger_number = num1 if num1 > num2 else num2
print(f"The larger number between {num1} and {num2} is {larger_number}")  # Expected output: 20

# Use the ternary operator to assign a value based on a condition
is_even = "Even" if num1 % 2 == 0 else "Odd"
print(f"{num1} is {is_even}")  # Expected output: 10 is Even
```
