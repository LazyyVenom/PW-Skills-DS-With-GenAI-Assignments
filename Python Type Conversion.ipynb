# Python Type Conversion

### Q1) Covert a integer to a floating-point number.

```python
number = 100
print(type(number),number) 
number = float(number)
print(type(number),number) 
```

### Q2) Covert a float to a integer.

```python
number = 100.50
print(type(number),number) 
number = int(number)
print(type(number),number) 
```

### Q3) Covert a integer to a String.

```python
number = 100
print(type(number),number) 
number_str = str(number)
print(type(number_str),number_str) 
```

### Q4) Covert a list to a tuple.

```python
my_list = [10,20,30,40,50]
print(type(my_list),my_list) 
my_tuple = tuple(my_list)
print(type(my_tuple),my_tuple) 
```

### Q5) Covert a tuple to a list.

```python
my_tuple = (10,20,30,40,50)
print(type(my_tuple),my_tuple) 
my_list = list(my_tuple)
print(type(my_list),my_list) 
```

### Q6) Covert a decimal number to binary.

```python
number = 10
print(bin(number))
print(bin(number)[2:])
```

### Q7) Covert a non-zero number to boolean.

```python
number = 10
print(bool(number))
```
